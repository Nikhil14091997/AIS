```python
def getChunks(text):
    ARTICLE = text
    ARTICLE = ARTICLE.replace('.', '.<eos>')
    ARTICLE = ARTICLE.replace('?', '?<eos>')
    ARTICLE = ARTICLE.replace('!', '!<eos>')
    sentences = ARTICLE.split('<eos>')
    #print(len(sentences))
    sentences = ARTICLE.split('<eos>')
    current_chunk = 0 
    chunks = []
    max_chunk = 500
    for sentence in sentences:
        if len(chunks) == current_chunk + 1: 
            if len(chunks[current_chunk]) + len(sentence.split(' ')) <= max_chunk:
                chunks[current_chunk].extend(sentence.split(' '))
            else:
                current_chunk += 1
                chunks.append(sentence.split(' '))
        else:
            chunks.append(sentence.split(' '))

    for chunk_id in range(len(chunks)):
        chunks[chunk_id] = ' '.join(chunks[chunk_id])
    return chunks
```

```python
def transformerLongSummary(data):
    '''
    Length of a general paragraph is 250 words.
    The length of data is not going to be greater than 500 words.
    If data is less than 250 words, we will not apply the summarization.
    If data is more than 250 words, we will apply the text summarization with
        max_length = 300 words
        min_length = 30% of words
    '''
    summary = []
    for chunk in data:
        words = len(chunk.split(' '))
        if words < 300:
            print("Content already short no need for summarization")
            summary.append(chunk)
        else:
            res = summarizer(chunk, max_length = 300, min_length = int(0.30*words), do_sample=False)
            summary.append(res)
    return summary
```

```python
longSummary = []
for text in df['content']:
    ret = getChunks(text)
    add = ''
    summary = transformerLongSummary(ret)
    for data in summary:
        partSummary = []
        if isinstance(data, list):
            for text in data:
                #print(text['summary_text'])
                partSummary.append(text['summary_text'])
        else:
            #print(data)
            partSummary.append(data)
        add = ''.join(elem for elem in partSummary)
    longSummary.append(add)
       
    
```

```python
df['Long-Summary-Transformer-Architecture'] = longSummary
```
